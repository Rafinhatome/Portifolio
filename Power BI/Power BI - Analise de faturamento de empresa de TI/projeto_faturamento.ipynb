```python
situacao = faturamento['Situação']
```

```python
clientes_ok = faturamento[faturamento['Situação'] == 'OK']
clientes_ok = clientes_ok[['Cliente', 'Valor bruto', 'Situação', 'Estado']]
clientes_ok
```

```python
clientes_ok_top10 = clientes_ok.groupby('Cliente').sum('Valor bruto').sort_values(by='Valor bruto', ascending=False)
clientes_ok_top10.head(10).plot(kind='bar')
```

```python
estados_ok = clientes_ok.groupby('Estado').sum('Valor bruto').sort_values(by='Valor bruto', ascending=False)
estados_ok.head(10).plot(kind='bar')
```

```python
clientes_vencida_top10 = clientes_vencida.groupby('Cliente').sum('Valor bruto').sort_values(by='Valor bruto', ascending=False)
clientes_vencida_top10.head(10).plot(kind='bar')
```
